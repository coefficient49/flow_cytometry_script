````python
from flowcyto_script import flowData as fD
from flowcyto_script import dataCollector as collector
from flowcyto_script import gates
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

sns.set(font_scale=2)  #big fonts
# folderDir = '/home/jeron/Documents/flow_data/20181207_HEK_COS/'
folderDir =input('What is the folder path to the fcs files? i.e. /home/jeron/Documents/flow_data/20181207_HEK_COS/\n\n')
if folderDir[-1] is not '/':
    folderDir = folderDir + '/'
filePaths_fcs = !ls $folderDir*.fcs
filePaths_csv = !ls $folderDir*sample*.csv
print('\n\nwe found these files:\n')
for x in filePaths_fcs:
    print(x)
````

## Analysis
Analyzing with gating (for custom gating use <fD_obj>.cGate(vertice))
Using the FD_Obj to first collect all the data, then analyse all the data using the threshold calculated using negative control.

````python
sampleDataEntry=pd.read_csv(filePaths_csv[0],delimiter=',')
df1, collection =collector(sampleDataEntry,filePaths_fcs)

## getting FITC gating threshold
keys = list(collection.keys())
fgating_value = [collection[x].Data for x in keys]
curated_data_appended=pd.Series([])
for x in range(3):
    data_to_analyse = fgating_value[x]['FITC-A']
    nonzero_data = data_to_analyse > 0
    curated_data = np.log10(data_to_analyse[nonzero_data])
    curated_data_appended=curated_data_appended.append(curated_data)
Y,X = np.histogram(curated_data_appended,bins=100)
cY = np.cumsum(Y)/np.sum(Y)
th = X[1::][cY > .996][0]
th = 10**th
df1, collection =collector(sampleDataEntry,filePaths_fcs,fgate=th)

df1.to_csv(folderDir+'flowData_out.csv')
print(df1)
plotMOI=False
if plotMOI:
    sns.boxplot('Virus','geoMean',data=df1)
    print('Analyzed data saved to \"{}flowData_out.csv\"'.format(folderDir))
    print('\nData preview:')
    print(df1.columns)
    plt.figure(figsize=(15,15))
    plt.subplot(2,2,1)
    sns.regplot(x='MOI',y='geoMean',data=df1)
    plt.subplot(2,2,2)
    sns.regplot(x='MOI',y='Percent +ve (%)',data=df1)
    plt.subplot(2,2,3)
    sns.regplot(x='MOI',y='TI',data=df1)
````
